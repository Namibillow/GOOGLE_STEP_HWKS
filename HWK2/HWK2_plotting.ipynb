```python
import matplotlib.pyplot as plt
import numpy as np
import sys
import time
%matplotlib inline
```

```python
times = []
ns = range(1,100,2)
for n in range(1, 100, 2):
    a = np.random.randint(40, size=(n, n))
    b = np.random.randint(40, size=(n, n))
    c = np.zeros((n, n))  # Matrix C
    start = time.time()
    
    # transpose b by switching row and col
    trans_b = [[row[i] for row in b] for i in range(n)]

    # store the sum of each row * col to its corresponding position in c
    c[:, :] = [[sum(x * y for x, y in zip(row_a, col_b))
                for col_b in trans_b] for row_a in a]

    end = time.time()
    times.append(end-start) 
```

```python
# plotting the points  
plt.plot(ns, times) 
  
# naming the x axis 
plt.xlabel('size of matrix') 
# naming the y axis 
plt.ylabel('execution time in seconds') 
  
# giving a title to my graph 
plt.title('relationship') 
  
# function to show the plot 
plt.show() 
```

```python

times = []
ns = range(1,100,2)
for n in range(1, 100, 2):
    a = np.random.randint(40, size=(n, n))
    b = np.random.randint(40, size=(n, n))
    c = np.zeros((n, n))  # Matrix C
    start = time.time()
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
#                 print(i,j,k)
                c[i][j] += a[i][k] * b[k][j]


    end = time.time()
    times.append(end-start) 
```

```python
# plotting the points  
plt.plot(ns, times) 
  
# naming the x axis 
plt.xlabel('size of matrix') 
# naming the y axis 
plt.ylabel('execution time in seconds') 
  
# giving a title to my graph 
plt.title('relationship (inefficient)') 
  
# function to show the plot 
plt.show() 
```
